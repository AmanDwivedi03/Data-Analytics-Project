## Data cleaning:-

As pressureInd coloumn have positive skew value, filling null attributes with mean value will be more appropriate.

```python
df.pressureInd.fillna(df.pressureInd.mean(),inplace=(True))
df.isnull().sum()
```

```python
df.fillna(df.median(),inplace=(True))
df.isnull().sum()
```

As 'moistureInd' have one unrealistic value it must be removed. using the following method:-

```python
df['moistureInd'].replace({df.moistureInd.max():df.moistureInd.mean()},inplace=True)
df.moistureInd.max()
```

## Univariate Analysis

```python
numerics=['lifetime', 'pressureInd', 'moistureInd', 'temperatureInd']
cat=['broken','team', 'provider']
```

```python
for col in numerics:
    plt.figure(figsize=(10,5))
    plt.scatter(np.arange(1000),df[col])
    plt.xlabel("Machines")
    plt.ylabel(col)
    plt.title(col)
    plt.show()
```

```python
for col in cat:
    plt.figure(figsize=(10,5))
   # plt.scatter(np.arange(10000),df[col])
    sns.countplot(df[col])
    #plt.xlabel("Customers")
    #plt.ylabel(col)
    plt.title(col)
    plt.show()
```

## BIVARIATE ANALYSIS

```python
plt.figure(figsize=(10,5))
sns.distplot(df.lifetime[df.broken==0])
sns.distplot(df.lifetime[df.broken==1])
plt.legend(['0','1'])
plt.show()
```

By analysing this graph we can say that:-

    Machines that have lifetime>60 have more probability of being broken.

## Multivariate analysis

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='team',y='lifetime',hue='broken',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='provider',y='lifetime',hue='broken',data=df)
plt.show()
```

Above graphs suggests that lifetime of the machine is the phenomenon that affects every Team as well as providers.

```python
plt.figure(figsize=(6,3))
sns.countplot(df['provider'])

plt.show()

plt.figure(figsize=(6,3))
sns.countplot(df['provider'][df.broken==1])
plt.show()
```

Above graph shows Provider 3 & 4 have most faulty machines.

```python
cor=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(cor, annot=True,cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df.moistureInd[df.broken==0])
sns.distplot(df.moistureInd[df.broken==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='moistureInd',y='lifetime',hue='broken',data=df)
plt.show()
```

Above graph suggests that Machines that have lifetime>60 have larger probability of being broken and Machines working at moistureind > 120 have lifetime >60 hence moisture may or may not be the factor affecting machines to be broken.

## Conclusion:-

From all above graphs & various analysis, it can be concluded that the lifetime of the Machine is an affecting factor why machines are getting damaged(broken). Its is also logical as the longer the machines have been working and running, higher are the chances of them to have wear and tear.
Machines with lifetime >60 have larger probablity of being broken hence adequate measures need to be taken for old/long run machines.
